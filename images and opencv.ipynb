```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('rgb.jpg')
```

```python
img.shape
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
pic=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
```

```python
pic.shape
```

```python
pic_resize=cv2.resize(pic,(200,400))
```

```python
plt.imshow(pic_resize)
```

```python
new_pic=cv2.resize(pic,(0,0),pic,0.5,0.5)
plt.imshow(new_pic)
```

```python
new_pic.shape
```

```python
pic_flip=cv2.flip(new_pic,0)
plt.imshow(pic_flip)
```

```python
pic_flip=cv2.flip(new_pic,1)
plt.imshow(pic_flip)
```

```python
 pic_flip=cv2.flip(new_pic,-1)
plt.imshow(pic_flip)
```

```python
cv2.imwrite('saved.jpg',new_pic)
```
