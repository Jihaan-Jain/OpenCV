```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=np.zeros(shape=(500,500),dtype=np.int32)
plt.imshow(img,cmap='grey')
```

```python
cv2.putText(img,"5",(100,400),cv2.FONT_HERSHEY_PLAIN,fontScale=30,color=(255,255,255),thickness=30)
plt.imshow(img,cmap='grey')
```

```python
cv2.putText(img,"Jihaan",(10,450),cv2.FONT_HERSHEY_PLAIN,fontScale=2.9,color=(255,255,255),thickness=15)
plt.imshow(img,cmap='grey')
```

```python
img1=np.zeros(shape=(500,500),dtype=np.int32)
plt.imshow(img1,cmap='grey')
```

```python
cv2.putText(img1,"5",(100,400),cv2.FONT_HERSHEY_PLAIN,fontScale=30,color=(255,255,255),thickness=30)
cv2.putText(img1,"5",(100,400),cv2.FONT_HERSHEY_PLAIN,fontScale=30,color=(0,0,0),thickness=15)
plt.imshow(img1,cmap='grey')
```

```python
cv2.putText(img1,"Jihaan",(10,450),cv2.FONT_HERSHEY_PLAIN,fontScale=2.9,color=(255,255,255),thickness=15)
cv2.putText(img1,"Jihaan",(10,450),cv2.FONT_HERSHEY_PLAIN,fontScale=2.9,color=(0,0,0),thickness=3)
plt.imshow(img1,cmap='grey')
```
