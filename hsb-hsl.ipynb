```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('rgb.jpg')
```

```python
plt.imshow(img)
```

```python
pic=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(pic)
```

```python
pic1=cv2.cvtColor(pic,cv2.COLOR_RGB2HSV)
plt.imshow(pic1)
```

```python
pic3=cv2.cvtColor(pic,cv2.COLOR_RGB2HLS)
plt.imshow(pic3)
```

```python
pic4=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(pic4)
```

```python
pic5=cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(pic5)
```
